## Relação dos candidatos eleitos de esquerda

### Deputados Federais

```python
#hide
 
import electionsBR as el

cand_elections_fed = {}
results_elections_fed = {}

position="Federal Deputy"

for year in [2014,2018]:
    
    cand = el.get_candidates(year=year, position=position,only_elected=True,state='RJ',
        columns=["NUMERO_CANDIDATO", "NOME_CANDIDATO", "SIGLA_UF", "NUM_TURNO", "SIGLA_PARTIDO", "COMPOSICAO_LEGENDA"])

    # partidos_esquerda = ['PT','PSOL','PC do B']
    cand = cand.query("SIGLA_UF == 'RJ' ")
    # cand = cand.query("SIGLA_UF == 'RJ' ").query(f"SIGLA_PARTIDO in {partidos_esquerda}")


    el_results = el.get_elections(year=year, position=position, state = "RJ", regional_aggregation="State", political_aggregation="Candidate",only_elected=True,
    columns=['NUMERO_CANDIDATO','QTDE_VOTOS'])
    el_results['QTDE_VOTOS'] = el_results['QTDE_VOTOS'].astype(int)
    votes_dict = el_results.set_index('NUMERO_CANDIDATO').to_dict()['QTDE_VOTOS']

    selected_cols = ['SIGLA_PARTIDO','NUMERO_CANDIDATO','NOME_CANDIDATO']
    cand = cand[selected_cols].copy()
    cand['QTDE_VOTOS'] = cand['NUMERO_CANDIDATO'].apply(lambda x: votes_dict.get(x))

    cand = cand.sort_values(by='QTDE_VOTOS',ascending=False)

    cand_elections_fed[year] = cand
    results_elections_fed[year] = el_results
```

```python
#hide

partidos_esquerda = ['PT','PSOL','PC do B']
hover_color = '#00BBBB'
styles = [dict(selector="tr:hover", props=[("background-color", f"{hover_color}")])]
```

> 2014 - Todos Partidos

```python
#hide_input
#collapse_output

year = 2014

df_styled = cand_elections_fed[year].style.apply(lambda x: ['background: #CC8888' if (set(partidos_esquerda).intersection(x.values)) 
                              else '' for i in x], 
                   axis=1)
df_styled.set_table_styles(styles)
```

> 2014 - Partidos de Esquerda

```python
#hide_input
#collapse_output

year = 2014
df_styled = cand_elections_fed[year].query(f'SIGLA_PARTIDO in {partidos_esquerda}').style

df_styled.set_table_styles(styles)
```

> 2018 - Todos Partidos

```python
#hide_input
#collapse_output

year=2018

df_styled = cand_elections_fed[year].style.apply(lambda x: ['background: #CC8888' if (set(partidos_esquerda).intersection(x.values)) 
                              else '' for i in x], 
                   axis=1)
df_styled.set_table_styles(styles)
```

> 2018 - Partidos de Esquerda

```python
#hide_input
#collapse_output

year = 2018
partidos_esquerda = ['PT','PSOL','PC do B']
# .query(f"SIGLA_PARTIDO in {partidos_esquerda}")

df_styled = cand_elections_fed[year].query(f"SIGLA_PARTIDO in {partidos_esquerda}").style

df_styled.set_table_styles(styles)
```
